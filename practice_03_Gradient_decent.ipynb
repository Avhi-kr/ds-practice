```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import math
```

```python
def gradient_descent(x,y):
  n=len(x)
  m_current=b_current=0
  iterations=1000000
  learning_rate=0.00002
  cost_previous=0
  for i in range(iterations):
    y_perdict=m_current*x +b_current
    cost=(1/n)* sum (value**2 for value in (y-y_perdict))#calculating the mean square error of the function
    md=-(2/n)*sum(x*(y-y_perdict))
    bd=-(2/n)*sum(y-y_perdict)
    m_current=m_current-learning_rate*md
    b_current=b_current-learning_rate*bd
    if math.isclose(cost,cost_previous,rel_tol=1e-20):
      break
    cost_previous=cost
  return m_current,b_current

```

```python
def gradient_descent_fromsklear(x,y):
  df=pd.read_csv("/content/test_scores (1).csv")
  model_train=LinearRegression()
  model_train.fit(df[['math']],df.cs)
  return model_train.coef_, model_train.intercept_
```

```python
df = pd.read_csv("/content/test_scores (1).csv")

x = np.array(df.math)
y = np.array(df.cs)

m, b = gradient_descent(x, y)
print("Using gradient descent:", m, b)

m_sklearn, b_sklearn = gradient_descent_fromsklear(x, y)
print("Using sklearn:", m_sklearn, b_sklearn)
```
